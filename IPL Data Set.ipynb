```python
import pandas as pd
ipl_data = pd.read_excel(file_path)
player_runs = ipl_data.groupby('batter')['batsman_runs'].sum().reset_index()
top_10_players = player_runs.nlargest(10, 'batsman_runs')
mean_runs = top_10_players['batsman_runs'].mean()
mode_runs = top_10_players['batsman_runs'].mode()
print(f"Mean of runs scored by top 10 players: {mean_runs}")
print(f"Mode of runs scored by top 10 players: {list(mode_runs)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "C:/Users/mynam/Desktop/ipl_matches_smalldata.xlsx"
ipl_data = pd.read_excel(file_path)
player_runs = ipl_data.groupby('batter')['batsman_runs'].sum().reset_index()
top_10_players = player_runs.nlargest(10, 'batsman_runs')
plt.figure(figsize=(10, 6))
plt.bar(top_10_players['batter'], top_10_players['batsman_runs'], color='skyblue')
plt.title('Top 10 Players by Total Runs', fontsize=16)
plt.xlabel('Player', fontsize=14)
plt.ylabel('Total Runs', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()
plt.show()
```
