```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = r"C:/Users/mynam/Desktop/tips.xlsx"
data = pd.read_excel(file_path)
print(data.head())
area_counts = data['Smoking'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(area_counts, labels=area_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Percentage of People in Smoking vs Non-Smoking')
plt.axis('equal') 
plt.show()
```
