```python
import cv2
image_path = r"C:/Users/mynam/Downloads/opencv/Bollywood-celebs-investing-ft-1-1-removebg-preview.png"
image = cv2.imread(image_path)
cv2.imshow("Original Image", image)
dimensions = image.shape
print(f"Original Image Dimensions: {dimensions} (Height, Width, Channels)")
resized_image = cv2.resize(image, (400, 256))
print("Resized Image Dimensions:", resized_image.shape)
cv2.imshow("Resized Image", resized_image)
flipped_image = cv2.flip(image, -1) 
cv2.imshow("Flipped Image", flipped_image) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
